```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sub_id = 'P010L'
```

```python
# Load the data
df = pd.read_csv(rf"E:\Gait Analysis NMH\CSV_raw\{sub_id}.csv", index_col='Frame')
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill missing values using linear interpolation
df_new = df.interpolate(method='cubicspline').round(2)
```

```python
df_new.to_csv(rf"E:\Gait Analysis NMH\CSV_Full\{sub_id}.csv")
```

```python
# Create a figure with two subplots vertically stacked
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Plot original data
ax1.plot(df)
ax1.set_title('Angle Before Interpolation')
ax1.set_xlabel('Frame')
ax1.set_ylabel('Angle')
ax1.grid(True)

# Plot interpolated data
ax2.plot(df_new)
ax2.set_title('Angle After Cubic Spline Interpolation')
ax2.set_xlabel('Frame')
ax2.set_ylabel('Angle')
ax2.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
